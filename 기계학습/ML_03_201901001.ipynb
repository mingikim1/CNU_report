```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/ML_Lab/report/2/lab2_test.csv")
df
```

```python
grouped = df.groupby('class')
group_mean = grouped.mean()
group_mean
```

```python
iris_setosa = df[df["class"] == 'Iris-setosa']
iris_versicolor =  df[df["class"] == 'Iris-versicolor']
iris_virginica =  df[df["class"] == 'Iris-virginica']
```

```python
plt.scatter(iris_setosa["petal_width"], iris_setosa["sepal_length"], color = 'r')
plt.scatter(iris_versicolor["petal_width"], iris_versicolor["sepal_length"], color = 'g')
plt.scatter(iris_virginica["petal_width"], iris_virginica["sepal_length"], color = 'b')
plt.xlabel("Petal_width")
plt.ylabel("Sepal_length")
plt.show()
```
